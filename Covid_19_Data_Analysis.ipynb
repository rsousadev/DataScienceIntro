## Mapa De Casos Confirmados Covid-19 3/11/20 - 3/12/20 - 3/13/20

```
fig = px.scatter_mapbox(covid_confirm, lat="Lat", lon="Long", hover_name="Country/Region",hover_data=["3/11/20","3/12/20","3/13/20", "3/14/20"],
                        color_discrete_sequence=["fuchsia"], zoom=3, height=300)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
